```python
from nba_api.stats.endpoints import franchiseplayers
```

```python
raps_players = franchiseplayers.FranchisePlayers(team_id=1610612761,per_mode_detailed='PerGame',season_type_all_star='Regular Season').get_data_frames()[0]
active_raps = raps_players[raps_players['ACTIVE_WITH_TEAM']==1]
active_raps.reset_index(inplace=True, drop=True)
```

```python
active_raps
```

```python
column_names = ['PLAYER','MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA','FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF','PTS', 'PLUS_MINUS']
games_won = pd.DataFrame(columns=column_names)
games_lost = pd.DataFrame(columns=column_names)
for i in range(0,len(active_raps['PERSON_ID'])):
  gamelog = playergamelog.PlayerGameLog(player_id=active_raps['PERSON_ID'][i],season_type_all_star='Regular Season').get_data_frames()[0]
  game_win = gamelog[gamelog['WL']=='W']
  game_loss = gamelog[gamelog['WL']=='L']
  game_win_columns = active_raps['PLAYER'][i], round(game_win['MIN'].mean(),3), round(game_win['FGM'].mean(),3), round(game_win['FGA'].mean(),3), round(game_win['FG_PCT'].mean(),3), round(game_win['FG3M'].mean(),3), round(game_win['FG3A'].mean(),3),	round(game_win['FG3_PCT'].mean(),3), round(game_win['FTM'].mean(),3), round(game_win['FTA'].mean(),3), round(game_win['FT_PCT'].mean(),3), round(game_win['OREB'].mean(),3), round(game_win['DREB'].mean(),3), round(game_win['REB'].mean(),3), round(game_win['AST'].mean(),3), round(game_win['STL'].mean(),3), round(game_win['BLK'].mean(),3), round(game_win['TOV'].mean(),3), round(game_win['PF'].mean(),3), round(game_win['PTS'].mean(),3), round(game_win['PLUS_MINUS'].mean(),3)
  game_loss_columns = active_raps['PLAYER'][i], round(game_loss['MIN'].mean(),3), round(game_loss['FGM'].mean(),3), round(game_loss['FGA'].mean(),3), round(game_loss['FG_PCT'].mean(),3), round(game_loss['FG3M'].mean(),3), round(game_loss['FG3A'].mean(),3),	round(game_loss['FG3_PCT'].mean(),3), round(game_loss['FTM'].mean(),3), round(game_loss['FTA'].mean(),3), round(game_loss['FT_PCT'].mean(),3), round(game_loss['OREB'].mean(),3), round(game_loss['DREB'].mean(),3), round(game_loss['REB'].mean(),3), round(game_loss['AST'].mean(),3), round(game_loss['STL'].mean(),3), round(game_loss['BLK'].mean(),3), round(game_loss['TOV'].mean(),3), round(game_loss['PF'].mean(),3), round(game_loss['PTS'].mean(),3), round(game_loss['PLUS_MINUS'].mean(),3)
  game_win_cols_list = list(game_win_columns)
  game_loss_cols_list = list(game_loss_columns)
  game_win_reshape = np.reshape(game_win_cols_list,(1,21))
  game_loss_reshape = np.reshape(game_loss_cols_list,(1,21))
  games_won = games_won.append(pd.DataFrame(game_win_reshape,columns=column_names))
  games_lost = games_lost.append(pd.DataFrame(game_loss_reshape,columns=column_names))
  time.sleep(.600)
games_won[['MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA','FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF',
           'PTS', 'PLUS_MINUS']] = games_won[['MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA','FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF','PTS', 'PLUS_MINUS']].apply(lambda col:pd.to_numeric(col,errors='coerce'))
games_lost[['MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA','FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF',
           'PTS', 'PLUS_MINUS']] = games_lost[['MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA','FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF','PTS', 'PLUS_MINUS']].apply(lambda col:pd.to_numeric(col,errors='coerce'))
```
