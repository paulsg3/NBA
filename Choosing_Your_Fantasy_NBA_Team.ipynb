```
def choose_my_team(df):
  center = df[df['Pos'].str.contains('C')].sort_values('YAHOO',ascending=False).head(2)
  PG1 = df['Pos'] == 'PG'
  PG2 = df['Pos'] == 'SG-PG'
  PG3 = df['Pos'] == 'PG-SG'
  SG1 = df['Pos'] == 'SG'
  SG2 = df['Pos'] == 'SG-SF'
  SG3 = df['Pos'] == 'SF-SG'
  SG4 = df['Pos'] == 'SG-PG'
  SG5 = df['Pos'] == 'PG-SG'
  guards = df.loc[PG1 | PG2 | PG3 | SG1 | SG2 | SG3 | SG4 | SG5].sort_values('YAHOO',ascending=False).head(3)
  forwards = df[df['Pos'].str.contains('F')].sort_values('YAHOO',ascending=False).head(3)
  util = df.sort_values('YAHOO',ascending=False).head(15)
  final_team = pd.concat([center, guards,forwards], axis=0)
  util = util.drop(util.index[util['Player'].isin(final_team['Player'])])
  #df = df.drop(df.index[df['colA'].isin([1.0, 2.0, 3.0])])
  final_team = pd.concat([center, guards,forwards,util], axis=0)
  #final_team = final_team.append(util[util.isin(final_team)==False])
  return final_team.sort_values(by='YAHOO',ascending=False)
```
